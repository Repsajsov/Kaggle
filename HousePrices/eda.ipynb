## MSSubClass


Dit zijn redelijk vage coderingen voor specifieke woningen. Ik heb in de metadata kijken naar wat iedere codering inhoudt. We kunnen nog meer informatie halen uit de coderingen omdat er vaak overlap zit. Zo zijn er meerdere coderingen die hetzelfde "aantal verdiepingen" bevat. Ik heb een dictionary gemaakt met de meest belangrijke en maak vervolgens extra dummy features.


```python
df.group_by("MSSubClass").len().sort("len", descending=True)
```

```python
sns.countplot(df, x="MSSubClass")
```

```python
df = df.with_columns(
    pl.when(pl.col("MSSubClass").is_in(value)).then(1).otherwise(0).alias(key) 
    for key, value in MSSUBCLASS_INFO.items()
)

df = df.to_dummies("MSSubClass")
```

```python
mssubclass_columns = [column for column in df.columns if column.startswith("MSSubClass_")] + [*MSSUBCLASS_INFO.keys()] + ["SalePrice"]
df[mssubclass_columns].sum() / len(df) * 100
```

```python
from scipy import stats

index = []
correlations = []
p_values = []
for column in mssubclass_columns:
    x = df[column].to_numpy()
    y = df["SalePrice"].to_numpy()
    
    correlation, p_value = stats.pointbiserialr(x, y)
    index.append(column)
    correlations.append(round(correlation, 2))
    p_values.append(round(p_value, 3))
    

corr_df = pl.DataFrame({
    "Variable": index,
    "Correlation": correlations,
    "PValue": p_values
})

corr_df = corr_df.filter((pl.col("PValue") < 0.05) & (pl.col("Correlation").abs() > 0.15)).sort("Correlation")
corr_df.sort("Correlation", descending=True)
```

```python
heatmap(df[mssubclass_columns])
```

```python
df.drop([variable for variable in mssubclass_columns if variable not in corr_df["Variable"]])
```

# Correlaties


```python

def heatmap(df, figsize=(20, 12), method="spearman"):
    corr_matrix = df.select(
        pl.col(pl.selectors.NUMERIC_DTYPES)
    ).to_pandas().corr(method=method)

    plt.figure(figsize=figsize)
    sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")

# columns = ["SalePrice"] + [*MSSUBCLASS_INFO.keys()]
# heatmap(df[columns])

# heatmap(df, figsize=(40,22))
```
